

Program to Represent Two-Dimensional Array Using Python List

```python
# Two-dimensional array
rows = int(input("Enter rows: "))
cols = int(input("Enter columns: "))

matrix = []

for i in range(rows):
    row = []
    for j in range(cols):
        num = int(input("Enter element: "))
        row.append(num)
    matrix.append(row)

print("Matrix:")
for i in matrix:
    print(i)
```
